```python
g,cluster_labels = cluster_keywords(data,cluster_preference=True,verbose=False)
draw_graph(data,g,cluster_labels)
```

```python
clusters = cluster_details(data,cluster_labels, plot=False, verbose=False)
```

```python
to_hide = ['transfert','lcl']
```

```python
fig = plt.figure(figsize=(18,12))
fig_args = {'fig':fig, 'shape':[(len(clusters)+4)/5,5],'sub':0,'title':'cluster'}
monthly_avg = []
sub_nb = 0
to_show = []
for c in clusters:
    fig_args['sub'] = sub_nb+1
    fig_args['title'] = c['name']
    group_idx = c['idx']
    group_avg = group_monthly_analysis(data, group_idx, start_month=3, fig_args = fig_args)
    if not(np.any([item in c['lab'] for item in to_hide])):
        monthly_avg.append(group_avg)
        to_show.append(sub_nb)
    sub_nb += 1

monthly_avg = np.asarray(monthly_avg)
plt.show()
```

```python
scaled_monthly_avg = monthly_avg / np.sum(monthly_avg,axis=0)
```

```python
# Expenses
plt.figure(figsize=(10,10))
plt.title('Monthly expenses')
pie(scaled_monthly_avg[:,1],
    labels=[clusters[to_show[l]]['name']+ '  ' + "{:.0f}".format(monthly_avg[l,1]) for l in arange(len(to_show))],
    colors = all_colors,
    autopct='%1.1f%%', shadow=True, startangle=90)
plt.legend(loc='center right', bbox_to_anchor=(1.5, 0.5))
plt.show()
```

```python
# Profits
plt.figure(figsize=(10,10))
plt.title('Monthly profits')
pie(scaled_monthly_avg[:,0],
    labels=[clusters[to_show[l]]['name']+ '  ' + "{:.0f}".format(monthly_avg[l,0]) for l in arange(len(to_show))],
    colors = all_colors,
    autopct='%1.1f%%', shadow=True, startangle=90)
plt.legend(loc='center right', bbox_to_anchor=(1.5, 0.5))
plt.show()
```
